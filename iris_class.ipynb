```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset
                                                   ['target'], random_state=0)
```

```python
print("X_train shape : {}".format(X_train.shape))
print("y train shape : {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
import numpy as np
iris_dataframe = pd.DataFrame(X_train, columns= iris_dataset.feature_names)
grr= pd.scatter_matrix(iris_dataframe, c=y_train, figsize= (15, 15), 
marker='o',hist_kwds={'bins': 20}, s=60, alpha =.8, cmap=mglearn.cm3)
```

K-Nearest Neighbors
decisionTree

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
#knn = KNeighborsClassifier(n_neighbors=1)
clf_tree = tree.DecisionTreeClassifier()
```

Building model
Use fit method
Takes two parameter(array, label)

```python
#knn.fit(X_train, y_train)
clf_tree = clf_tree.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9,1,0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

```python
with open("iris.dot", 'w') as f:
    f = tree.export_graphviz(clf_tree, out_file=f)
```

```python
import os
os.unlink('iris.dot')
```

```python
import pydotplus
clf_tree = clf_tree.fit(iris_dataset.data, iris_dataset.target)
dot_data = tree.export_graphviz(clf_tree, out_file=None) 
graph = pydotplus.graph_from_dot_data(dot_data) 
graph.write_pdf("iris.pdf") 
```

```python
from IPython.display import Image  
dot_data = tree.export_graphviz(clf_tree, out_file=None, 
                         feature_names=iris_dataset.feature_names,  
                         class_names=iris_dataset.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```

```python
Making predictions
```

```python
#prediction = knn.predict(X_new)
prediction =clf_tree.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name:{}".format(iris_dataset['target_names'][prediction]))

predict1=clf_tree.predict([[4.7,3.2,1.3,0.2]])
print(predict1)
```

Evaluate the Model

```python
#y_pred = knn.predict(X_test)
y_pred = clf_tree.predict(X_test)
print("Test set prediction: \n {}".format(y_pred))
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred==y_test)))
```

```python
#print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
print("Test set score: {:.2f}".format(clf_tree.score(X_test, y_test)))
```
